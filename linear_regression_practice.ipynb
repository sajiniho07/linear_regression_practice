```python
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from sklearn.linear_model import LinearRegression

ws = load_workbook("/linear_regression_data.xlsx")["Sheet1"]
max_row = ws.max_row

data = []
for i in range(2, max_row + 1):
    data.append([ws[f"A{i}"].value, ws[f"B{i}"].value, ws[f"C{i}"].value, ws[f"D{i}"].value])

data = np.array(data)
input_1 = data[:, 0]
input_2 = data[:, 1]
input_3 = data[:, 2]
output = data[:, 3]

def model_prediction(x, y):
  model = LinearRegression()
  x = x.reshape((-1,1))
  y = y.reshape((-1,1))

  model.fit(x, y)
  yprd = model.predict(x)

  plt.subplot(2, 1, 1)
  plt.plot(x, y, "o", label="x & y")
  plt.plot(x, yprd, "o", label="x & yprd")
  plt.legend()
  return yprd

def plot_output_prediction(output, prediction):
  plt.subplot(2, 1, 2)
  plt.plot(output, prediction, "o")
  plt.xlabel("output")
  plt.ylabel("prediction")
```

```python

yprd = model_prediction(input_1, output)
plot_output_prediction(output, yprd)
```

```python

yprd = model_prediction(input_2, output)
plot_output_prediction(output, yprd)
```

```python

yprd = model_prediction(input_3, output)
plot_output_prediction(output, yprd)
```
